#1. Análisis de facturación



## Descripción

En este primer análisis, debes calcular el **ingreso total** de cada tienda. **Sumando los valores de la columna Precio** de cada conjunto de datos de la tienda para estimar los ingresos.

```python
# Calcular ingresos
ingreso_tienda = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Imprimir resultados
print(f"Ingreso total de la tienda: ${ingreso_tienda:,.2f}")
print(f"Ingreso total de la tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso total de la tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso total de la tienda 4: ${ingreso_tienda4:,.2f}")

# Crear diccionario con ingresos
ingresos = {
    "tienda": ingreso_tienda,
    "tienda2": ingreso_tienda2,
    "tienda3": ingreso_tienda3,
    "tienda4": ingreso_tienda4
}

# Identificar tienda con mayor ingreso
tienda_mayor_ingreso = max(ingresos, key=ingresos.get)
print(f"\nLa tienda con el mayor ingreso es: {tienda_mayor_ingreso} con ${ingresos[tienda_mayor_ingreso]:,.2f}")

```

```python
import matplotlib.pyplot as plt

# Datos
nombres_tiendas = list(ingresos.keys())
valores_ingresos = list(ingresos.values())

# Gráfico
plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, valores_ingresos, color='skyblue')
plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos ($)")
plt.tight_layout()
plt.show()
```

#### Análisis (1)

La tienda con mayores ingresos totales es la **Tienda 1**, seguida muy de cerca por la **Tienda 2** y la **Tienda 3**, con diferencias pequeñas entre ellas. La **Tienda 4**, sin embargo, muestra una caída más marcada en ingresos. Esto podría deberse a un menor volumen de ventas, precios más bajos, o menor popularidad. Aunque todas generan ingresos elevados, la **Tienda 4** se destaca como la menos rentable del grupo.

Sin embargo, la diferencia entre los ingresos de cada tienda no es tan grande como para poder determinar, a partir de esos datos, cuál tienda debería vender el señor Juan.

# 2. Ventas por categoría

## Descripción

En este debes **calcular la cantidad de productos vendidos por categoría** en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando **las categorías más populares** de cada tienda.

```python
# Calcular cantidad de productos vendidos por categoría
productos_por_categoria_tienda = tienda["Categoría del Producto"].value_counts()
productos_por_categoria_tienda2 = tienda2["Categoría del Producto"].value_counts()
productos_por_categoria_tienda3 = tienda3["Categoría del Producto"].value_counts()
productos_por_categoria_tienda4 = tienda4["Categoría del Producto"].value_counts()

# Imprimir resultados
print("VENTAS POR CATEGORÍA:\n")
print("Tienda:\n")
print(productos_por_categoria_tienda)
print("\nTienda 2:\n")
print(productos_por_categoria_tienda2)
print("\nTienda 3:\n")
print(productos_por_categoria_tienda3)
print("\nTienda 4:\n")
print(productos_por_categoria_tienda4)

# Extraer categorías más vendidas
categoria_mas_vendida_tienda = tienda["Categoría del Producto"].value_counts().idxmax()
cantidad_mas_vendida_tienda = tienda["Categoría del Producto"].value_counts().max()

categoria_mas_vendida_tienda2 = tienda2["Categoría del Producto"].value_counts().idxmax()
cantidad_mas_vendida_tienda2 = tienda2["Categoría del Producto"].value_counts().max()

categoria_mas_vendida_tienda3 = tienda3["Categoría del Producto"].value_counts().idxmax()
cantidad_mas_vendida_tienda3 = tienda3["Categoría del Producto"].value_counts().max()

categoria_mas_vendida_tienda4 = tienda4["Categoría del Producto"].value_counts().idxmax()
cantidad_mas_vendida_tienda4 = tienda4["Categoría del Producto"].value_counts().max()

# Imprimir resultados
print(f"\nLa categoría más vendida en la tienda es: {categoria_mas_vendida_tienda} con {cantidad_mas_vendida_tienda} productos vendidos.")
print(f"La categoría más vendida en la tienda 2 es: {categoria_mas_vendida_tienda2} con {cantidad_mas_vendida_tienda2} productos vendidos.")
print(f"La categoría más vendida en la tienda 3 es: {categoria_mas_vendida_tienda3} con {cantidad_mas_vendida_tienda3} productos vendidos.")
print(f"La categoría más vendida en la tienda 4 es: {categoria_mas_vendida_tienda4} con {cantidad_mas_vendida_tienda4} productos vendidos.")
```

```python
# Sumar categorías entre todas
categorias_combinadas = (
    tienda["Categoría del Producto"]
    .value_counts()
    .add(tienda2["Categoría del Producto"].value_counts(), fill_value=0)
    .add(tienda3["Categoría del Producto"].value_counts(), fill_value=0)
    .add(tienda4["Categoría del Producto"].value_counts(), fill_value=0)
)

top_categorias = categorias_combinadas.sort_values(ascending=False).head(5).index

# Crear un DataFrame para graficar
df_categorias = pd.DataFrame({
    "Tienda 1": tienda["Categoría del Producto"].value_counts(),
    "Tienda 2": tienda2["Categoría del Producto"].value_counts(),
    "Tienda 3": tienda3["Categoría del Producto"].value_counts(),
    "Tienda 4": tienda4["Categoría del Producto"].value_counts()
}).fillna(0).loc[top_categorias]

# Gráfico de barras agrupadas
df_categorias.plot(kind='bar', figsize=(10, 6))
plt.title("Top 5 Categorías de Productos Vendidos por Tienda")
plt.xlabel("Categoría")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Análisis (2)

**Muebles** es, con claridad, la categoría más vendida en las cuatro tiendas, superando las 440 unidades en cada una, lo que indica una demanda alta y constante.

**Electrónicos** sigue como la segunda categoría más popular, también con ventas fuertes y bastante parejas entre tiendas.

Las categorías de **Juguetes, Electrodomésticos y Deportes y diversión** completan el top 5, aunque con diferencias más notables: por ejemplo, tienda 4 tiene más juguetes vendidos que el resto, pero menos electrodomésticos.

Las tiendas son consistentes en los tipos de productos más vendidos, lo que sugiere una similitud en el perfil de sus clientes o en las estrategias de venta.

Aunque el análisis de las **categorías más vendidas** nos permite entender qué tipo de productos tienen mayor salida en cada tienda, esta información **no es suficiente** por sí sola para decidir cuál tienda debe vender el señor Juan. Las ventas por categoría reflejan popularidad, pero no consideran la rentabilidad, el costo operativo, la satisfacción del cliente o los márgenes de ganancia por categoría. Además, el hecho de que todas las tiendas tengan un patrón de ventas similar en categorías populares sugiere que eliminar una tienda podría afectar de forma similar a la cadena entera. Es necesario complementar este análisis con otros factores como ingresos, reseñas de clientes, costos de envío y productos más/menos vendidos antes de tomar una decisión.

# 3. Calificación promedio de la tienda


## Descripción

En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es **conocer la satisfacción del cliente con los productos vendidos**.

```python
# Calcular calificación promedio
calificacion_promedio_tienda = tienda["Calificación"].mean()
calificacion_promedio_tienda2 = tienda2["Calificación"].mean()
calificacion_promedio_tienda3 = tienda3["Calificación"].mean()
calificacion_promedio_tienda4 = tienda4["Calificación"].mean()

# Imprimir resultados
print(f"Calificación promedio de la tienda 1: {calificacion_promedio_tienda:.2f}")
print(f"Calificación promedio de la tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación promedio de la tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación promedio de la tienda 4: {calificacion_promedio_tienda4:.2f}")

# Crear diccionario con calificaciones promedio
calificaciones = {
    "tienda1": calificacion_promedio_tienda,
    "tienda2": calificacion_promedio_tienda2,
    "tienda3": calificacion_promedio_tienda3,
    "tienda4": calificacion_promedio_tienda4
}

# Identificar tienda con mayor y menor calificación
tienda_mejor_calificada = max(calificaciones, key=calificaciones.get)
tienda_peor_calificada = min(calificaciones, key=calificaciones.get)

# Imprimir resultados
print(f"\nLa tienda con mayor calificación promedio es: {tienda_mejor_calificada} con {calificaciones[tienda_mejor_calificada]:.2f}")
print(f"La tienda con menor calificación promedio es: {tienda_peor_calificada} con {calificaciones[tienda_peor_calificada]:.2f}")
```

#### Análisis (3)

El análisis de las **calificaciones promedio** revela que todas las tiendas tienen valoraciones bastante cercanas, oscilando entre **3.98 y 4.05**, lo cual indica una satisfacción del cliente relativamente alta y uniforme. La **tienda 3** destaca ligeramente como la mejor evaluada, mientras que la **tienda 1** tiene la calificación más baja, aunque la diferencia es mínima. Esto sugiere que **la experiencia del cliente no varía drásticamente entre tiendas**, por lo que esta métrica **no es decisiva por sí sola** para justificar el cierre de alguna sucursal.

En conclusión, aunque las calificaciones pueden ser un indicador útil de desempeño, en este caso **no hay una tienda claramente deficiente** desde la perspectiva del cliente. Será necesario cruzar esta información con otros indicadores, como ingresos y costos, para tomar una decisión sólida.

```python
# Calcular la cantidad de cada producto que se ha vendido
ventas_productos_tienda = tienda["Producto"].value_counts()
ventas_productos_tienda2 = tienda2["Producto"].value_counts()
ventas_productos_tienda3 = tienda3["Producto"].value_counts()
ventas_productos_tienda4 = tienda4["Producto"].value_counts()

# Imprimir resultados
print("VENTAS POR PRODUCTO:\n")
print("Tienda:\n")
print(ventas_productos_tienda)
print("\nTienda 2:\n")
print(ventas_productos_tienda2)
print("\nTienda 3:\n")
print(ventas_productos_tienda3)
print("\nTienda 4:\n")
print(ventas_productos_tienda4)

# Extraer productos más vendidos
producto_mas_vendido_tienda = ventas_productos_tienda.idxmax()
cantidad_mas_vendido_tienda = ventas_productos_tienda.max()
producto_menos_vendido_tienda = ventas_productos_tienda.idxmin()
cantidad_menos_vendido_tienda = ventas_productos_tienda.min()

producto_mas_vendido_tienda2 = ventas_productos_tienda2.idxmax()
cantidad_mas_vendido_tienda2 = ventas_productos_tienda2.max()
producto_menos_vendido_tienda2 = ventas_productos_tienda2.idxmin()
cantidad_menos_vendido_tienda2 = ventas_productos_tienda2.min()

producto_mas_vendido_tienda3 = ventas_productos_tienda3.idxmax()
cantidad_mas_vendido_tienda3 = ventas_productos_tienda3.max()
producto_menos_vendido_tienda3 = ventas_productos_tienda3.idxmin()
cantidad_menos_vendido_tienda3 = ventas_productos_tienda3.min()

producto_mas_vendido_tienda4 = ventas_productos_tienda4.idxmax()
cantidad_mas_vendido_tienda4 = ventas_productos_tienda4.max()
producto_menos_vendido_tienda4 = ventas_productos_tienda4.idxmin()
cantidad_menos_vendido_tienda4 = ventas_productos_tienda4.min()

# Imprimir resultados
print(f"\nProducto más vendido en tienda 1: {producto_mas_vendido_tienda} ({cantidad_mas_vendido_tienda} unidades)")
print(f"Producto menos vendido en tienda 1: {producto_menos_vendido_tienda} ({cantidad_menos_vendido_tienda} unidades)")

print(f"Producto más vendido en tienda 2: {producto_mas_vendido_tienda2} ({cantidad_mas_vendido_tienda2} unidades)")
print(f"Producto menos vendido en tienda 2: {producto_menos_vendido_tienda2} ({cantidad_menos_vendido_tienda2} unidades)")

print(f"Producto más vendido en tienda 3: {producto_mas_vendido_tienda3} ({cantidad_mas_vendido_tienda3} unidades)")
print(f"Producto menos vendido en tienda 3: {producto_menos_vendido_tienda3} ({cantidad_menos_vendido_tienda3} unidades)")

print(f"Producto más vendido en tienda 4: {producto_mas_vendido_tienda4} ({cantidad_mas_vendido_tienda4} unidades)")
print(f"Producto menos vendido en tienda 4: {producto_menos_vendido_tienda4} ({cantidad_menos_vendido_tienda4} unidades)")
```

Con base en los productos más y menos vendidos por tienda, se observa que **cada tienda tiene un rendimiento relativamente equilibrado**, sin que alguna esté fallando drásticamente en su catálogo. # 5. Envío promedio por tienda

## Descripción

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
# Calcular costo de envío promedio por tienda
envio_promedio_tienda = tienda["Costo de envío"].mean()
envio_promedio_tienda2 = tienda2["Costo de envío"].mean()
envio_promedio_tienda3 = tienda3["Costo de envío"].mean()
envio_promedio_tienda4 = tienda4["Costo de envío"].mean()

# Mostrar resultados
print(f"Costo de envío promedio en Tienda 1: ${envio_promedio_tienda:,.2f}")
print(f"Costo de envío promedio en Tienda 2: ${envio_promedio_tienda2:,.2f}")
print(f"Costo de envío promedio en Tienda 3: ${envio_promedio_tienda3:,.2f}")
print(f"Costo de envío promedio en Tienda 4: ${envio_promedio_tienda4:,.2f}")

# Crear diccionario con el costo promedio de envío
costos_envio = {
    "Tienda 1": envio_promedio_tienda,
    "Tienda 2": envio_promedio_tienda2,
    "Tienda 3": envio_promedio_tienda3,
    "Tienda 4": envio_promedio_tienda4
    }

# Identificar tiendas con mayor y menos costo promedio de envío
tienda_envio_mas_caro = max(costos_envio, key=costos_envio.get)
tienda_envio_mas_barato = min(costos_envio, key=costos_envio.get)

print(f"\nLa tienda con el envío más caro es: {tienda_envio_mas_caro} (${costos_envio[tienda_envio_mas_caro]:,.2f})")
print(f"La tienda con el envío más barato es: {tienda_envio_mas_barato} (${costos_envio[tienda_envio_mas_barato]:,.2f})")


```

```python
import matplotlib.pyplot as plt

# Datos
tiendas = list(costos_envio.keys())
costos = list(costos_envio.values())

# Crear gráfico de dispersión
plt.figure(figsize=(8, 5))
plt.scatter(tiendas, costos, color='green', s=100)  # s= tamaño de los puntos

# Añadir valores encima de cada punto
for i, costo in enumerate(costos):
    plt.text(tiendas[i], costo + 0.2, f"${costo:.2f}", ha='center')

# Estética
plt.title("Costo Promedio de Envío por Tienda (Gráfico de Dispersión)")
plt.xlabel("Tienda")
plt.ylabel("Costo Promedio ($)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

#### Análisis (5)

Con los datos de **costos de envío promedio**, ahora contamos con una dimensión adicional importante: la eficiencia logística. La **tienda 1** no solo tiene el **costo de envío más alto ($26,018.61)**, sino que también tiene la **calificación promedio más baja (3.98)**. Aunque fue la que generó **mayores ingresos**, estos costos más altos y una satisfacción ligeramente menor podrían comprometer su rentabilidad a largo plazo.

Por otro lado, la **tienda 4** tiene el envío más barato ($23,459.46) y una calificación promedio aceptable (4.00), aunque sus ingresos fueron más bajos que otras tiendas. Esto sugiere que, con una estrategia adecuada, podría volverse más competitiva y rentable.

# Extra: Análisis del desempeño geográfico

En este extra, tendrás el desafío de explorar las coordenadas geográficas de los datos de ventas e identificar patrones relacionados con la ubicación de las compras. Al utilizar las columnas de latitud y longitud, puede generar visualizaciones para comprender cómo varían las ventas según la ubicación geográfica.

```python
# Añadir identificador de tienda
tienda["tienda"] = "Tienda 1"
tienda2["tienda"] = "Tienda 2"
tienda3["tienda"] = "Tienda 3"
tienda4["tienda"] = "Tienda 4"

# Concatenar en un solo DataFrame
ventas_geo = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo visual
sns.set(style="whitegrid")

# Tamaño del gráfico
plt.figure(figsize=(12, 8))

# Gráfico de dispersión por ubicación, coloreado por tienda
sns.scatterplot(data=ventas_geo, x="lon", y="lat", hue="tienda", alpha=0.6)

# Título y etiquetas
plt.title("Distribución geográfica de ventas por tienda", fontsize=16)
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend(title="Tienda")
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Mapa de calor de densidad
sns.kdeplot(
    data=ventas_geo,
    x="lon",
    y="lat",
    fill=True,
    cmap="Reds",
    bw_adjust=0.5,
    levels=100,
    alpha=0.6
)

# Título
plt.title("Mapa de calor de densidad de ventas", fontsize=16)
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.show()
```

```python
import folium
from folium.plugins import HeatMap

# Crear un mapa base centrado en el promedio de coordenadas
mapa = folium.Map(location=[ventas_geo["lat"].mean(), ventas_geo["lon"].mean()], zoom_start=6)

# Crear datos para HeatMap
heat_data = [[row["lat"], row["lon"]] for index, row in ventas_geo.iterrows()]

# Añadir HeatMap
HeatMap(heat_data).add_to(mapa)

# Mostrar mapa
mapa
```

Ninguna tienda parece tener dominio exclusivo en una región, lo cual puede deberse a que todas comparten la misma red de distribución.
